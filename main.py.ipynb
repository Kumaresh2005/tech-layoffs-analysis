```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Create output folder
os.makedirs("output", exist_ok=True)
```

```python
# Load dataset
df = pd.read_csv("tech_layoffs_hiring_trends_elite_v2.csv")
```

```python
print("Original Shape:", df.shape)
```

### DATA CLEANING


```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Remove extra spaces from column names
df.columns = df.columns.str.strip()
```

```python
# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
# Standardize text columns
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.title()

print("Cleaned Shape:", df.shape)
```

### SAVE CLEANED DATA

```python
cleaned_file = "output/cleaned_data.csv"
df.to_csv(cleaned_file, index=False)
```

```python
print(f"Cleaned dataset saved at: {cleaned_file}")
```

### GENERATE SUMMARY REPORT

```python
summary = df.describe(include='all')
```

```python
summary.to_csv("output/summary.csv")
```

```python
with open("output/report.txt", "w") as f:
    f.write("TECH LAYOFFS & HIRING REPORT\n")
    f.write("="*40 + "\n\n")
    f.write(str(summary))
```

### LAYOFF ANALYSIS

```python
numeric_cols = df.select_dtypes(include='number').columns

if len(numeric_cols) > 0:

    main_col = numeric_cols[0]

    top_data = df.groupby(df.columns[0])[main_col].sum().sort_values(ascending=False).head(10)

    # Layoff Chart
    plt.figure(figsize=(10,6))

    top_data.plot(kind="bar")

    plt.title("Top 10 Trends Analysis")
    plt.xlabel(df.columns[0])
    plt.ylabel(main_col)

    plt.xticks(rotation=45)

    plt.tight_layout()

    plt.savefig("output/layoffs_chart.png")

    plt.close()

print("Project Completed Successfully")
```
